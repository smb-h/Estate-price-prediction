```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_excel('../data/raw/main_feats.xlsx')
df.head()
```

```python
df.rename(columns={'Srat Quarter': 'start_quarter'}, inplace=True)
df.rename(columns={'Complition Quarter': 'end_quarter'}, inplace=True)
df.rename(columns={'Start year': 'start_year'}, inplace=True)
df.rename(columns={'Compliton  year': 'end_year'}, inplace=True)
df.rename(columns={'X-1': 'x1'}, inplace=True)
df.rename(columns={'X-2': 'x2'}, inplace=True)
df.rename(columns={'X-3': 'x3'}, inplace=True)
df.rename(columns={'X-5': 'x5'}, inplace=True)
df.rename(columns={'X-5.1': 'x51'}, inplace=True)
df.rename(columns={'cost/m^2': 'y'}, inplace=True)
df.head()
```

```python
# reorder columns to start_year, start_quarter, end_year, end_quarter, x1, x2, x3, x5, x51, y
df = df[['start_year', 'start_quarter', 'end_year', 'end_quarter', 'x1', 'x2', 'x3', 'x5', 'x51', 'y']]
df.head()
```

```python
df['start_year'] = df['start_year'].apply(lambda x: x + 1300 if x < 1400 else x)
df['end_year'] = df['end_year'].apply(lambda x: x + 1300 if x < 1400 else x)
df.head()
```

```python
df["project_duration"] = ((df['end_year'] - df['start_year']) * 12) + ((df['end_quarter'] - df['start_quarter']) * 3)
df.tail()
```

```python
def calculate_date(start_year, start_quarter):
    start_year = int(start_year)
    start_quarter = int(start_quarter)
    if start_quarter == 1:
        return str(start_year) + "-" + "01" + "-" + "01" 
    elif start_quarter == 2:
        return str(start_year) + "-" + "04" + "-" + "01" 
    elif start_quarter == 3:
        return str(start_year) + "-" + "07" + "-" + "01" 
    elif start_quarter == 4:
        return str(start_year) + "-" + "10" + "-" + "01" 
```

```python
df['start_date'] = df.apply(lambda row: calculate_date(row['start_year'], row['start_quarter']), axis=1)
df['end_date'] = df.apply(lambda row: calculate_date(row['end_year'], row['end_quarter']), axis=1)
df.tail()
```

```python
df = df[['start_date', 'start_year', 'start_quarter', 'end_date', 'end_year', 'end_quarter', 'project_duration', 'x1', 'x2', 'x3', 'x5', 'x51', 'y']]
df.tail()
```

```python
# drop start_year, start_quarter, end_year, end_quarter
df = df.drop(['start_year', 'start_quarter', 'end_year', 'end_quarter'], axis=1)
df.tail()
```

```python
# save to csv
df.to_csv('../data/processed/main_feats.csv', index=False)
```
